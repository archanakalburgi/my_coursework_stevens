Solution:
Adjusted the weights till the class labels were correctly classified 

Explaination:
- The labels were correctly classified after 91 iterations, i.e when number of errors(Misclassifications) becomes zero

- The same can be verified in the further iteration, where there is no difference in the weight vectors and error(Misclassifications) = 0 for all further iteration

- I have performed 100 iterations. 

```python
import numpy as np

learning_rate = 0.01
no_iters = 10000 

data = np.array([
                [ 1, 1, -1,  0, 2,  1],
                [ 0, 0,  1,  2, 0, -1],
                [-1,-1,  1  ,1, 1,  1],
                [4,  0,  1,  2, 1, -1],
                [-1, 0,  1,  2, 1, -1],
                [-1, -1, -1, 1, 0, -1],
                [-1, 1,   1, 2, 1,  1]
                ])
labels = np.array([1, -1, 1, -1, -1, -1, 1])

weights = np.array([3.0, 1, 1, -1, 2, -7])
errs = []
iteration = 1 

# Perceptron implementation 

for _ in range(1,101):
  err = 0 
  print(f"Iteration: {iteration}")
  print(f"Weights Vector: {weights}") 
  iteration += 1
  for xi, target in zip(data, labels):
    res = np.dot(xi, weights) + weights[0]
    correction = np.where(res >= 0.0, 1, -1)
    update = learning_rate *  (target - correction)
    weights += update * xi 
    weights[0] += update
    err += int(update != 0.0)
    errs.append(err)
  print(f"Misclassifications: {err}") 
  print("\n")
#   print(weights) 
```

The labels were correctly classified after 90 iterations
