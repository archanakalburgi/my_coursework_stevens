Repeated the same process for k = 5 and k = 11

*Note : Only those libraries essential to work with data frames are imported 

```python
# importing necessary libraries
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split 
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv("/content/drive/MyDrive/ML_Assignment2/pima-indians-diabetes.csv")

data = df.to_numpy()
X = data[:, [1,2,3,8]]

```

```python
import math
# distance calculation 
def distance(row1, row2):
    dist = 0
    for i in range(len(row1)-1): # -1 as we do not want checking output distance
        dist = dist + (row1[i]-row2[i])**2
    return math.sqrt(dist)

def cls_with_k_nghs(test, train, k): # row -> class(0/1) 
    dists = list(map(lambda x : (distance(x, test), x), train))
    dists.sort(key = lambda x : x[0])
    classes = np.array(list(map(lambda x : x[-1][-1], dists[0:k])))
    no_ones = np.sum(classes==1)
    no_zeros = np.sum(classes==0)
    predicted_cls = 1 if no_ones > no_zeros else 0
    return test[-1] == predicted_cls

def test_predict_and_score(X, k, test): # prediction accuracy 
    xlss = np.array(list(map(lambda row: cls_with_k_nghs(row, X, k), test)))
    no_true = len(list(filter(lambda x : x == True, xlss)))
    return no_true/len(X)

accuracy_1= []
accuracy_5 = []
accuracy_11 =[]

for k in [1,5,11]:
    for i in range(11):
        X_train, X_test= train_test_split(X, test_size=0.5) 
        accuracy_1.append(test_predict_and_score(X_train, 1, X_test))
        accuracy_5.append(test_predict_and_score(X_train, 5, X_test))
        accuracy_11.append(test_predict_and_score(X_train, 11, X_test))

print(f" Mean accuracy for k = 1 is {np.array(accuracy_1).mean()}")
print(f" Standard Deviation for k = 1 is {np.array(accuracy_1).std()}")
print("\n")
print(f" Mean accuracy for k = 5 is {np.array(accuracy_5).mean()}")
print(f" Standard Deviation for k = 5 is {np.array(accuracy_5).std()}")
print("\n")
print(f" Mean accuracy for k = 11 is {np.array(accuracy_11).mean()}")
print(f" Standard Deviation for k = 11 is {np.array(accuracy_11).std()}")
```
